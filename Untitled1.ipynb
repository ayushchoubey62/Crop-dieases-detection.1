```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define training and validation data generators
train_datagen = ImageDataGenerator(
    rescale=1.0/255, 
    rotation_range=20, 
    width_shift_range=0.2, 
    height_shift_range=0.2, 
    shear_range=0.2, 
    zoom_range=0.2, 
    horizontal_flip=True,
    validation_split=0.2  # 20% data for validation
)

# Load training and validation data
train_generator = train_datagen.flow_from_directory(
    r'C:\Users\ayush\OneDrive\Desktop\project 1\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\train',  # Replace with the correct dataset path
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    r'C:\Users\ayush\OneDrive\Desktop\project 1\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\valid',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, Input

model = models.Sequential([
    Input(shape=(150, 150, 3)),  # Define input explicitly
    layers.Conv2D(32, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(512, activation="relu"),
    layers.Dense(38, activation="softmax")
])

model.summary()
```

```python
model.compile(
    optimizer="adam",  
    loss="categorical_crossentropy",  # Use "sparse_categorical_crossentropy" if labels are not one-hot encoded
    metrics=["accuracy"]
)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5,  # Increase epochs if needed
    verbose=1
)
```

```python
import tkinter as tk
from tkinter import filedialog, messagebox, ttk, Toplevel
from PIL import Image, ImageTk, ImageEnhance, ImageFilter
import threading
import time
import os
from fpdf import FPDF
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

class App(tk.Tk):
    def __init__(self):
        super().__init__()

        # Load the model
        self.model = load_model("my_model.keras")

        # Define class labels - make sure these match your model's output
        self.class_labels = {'Apple_Apple_scab': 0, 'Apple_Black_rot': 1, 'Apple_Cedar_apple_rust': 2, 'Apple_healthy': 3, 'Blueberry_healthy': 4, 'Cherry_Powdery_mildew': 5, 'Cherry_healthy': 6, 'Corn_Cercospora_leaf_spot_Gray_leaf_spot': 7, 'Corn_Common_rust': 8, 'Corn_Northern_Leaf_Blight': 9, 'Corn_healthy': 10, 'Grape_Black_rot': 11, 'Grape_Esca_Black_Measles': 12, 'Grape_Leaf_blight': 13, 'Grape_healthy': 14, 'Orange_Haunglongbing_Citrus_greening': 15, 'Peach_Bacterial_spot': 16, 'Peach_healthy': 17, 'Pepper_bell_Bacterial_spot': 18, 'Pepper_bell_healthy': 19, 'Potato_Early_blight': 20, 'Potato_Late_blight': 21, 'Potato_healthy': 22, 'Raspberry_healthy': 23, 'Soybean_healthy': 24, 'Squash_Powdery_mildew': 25, 'Strawberry_Leaf_scorch': 26, 'Strawberry_healthy': 27, 'Tomato_Bacterial_spot': 28, 'Tomato_Early_blight': 29, 'Tomato_Late_blight': 30, 'Tomato_Leaf_Mold': 31, 'Tomato_Septoria_leaf_spot': 32, 'Tomato_Spider_mites _Two_spotted_spider_mite': 33, 'Tomato_Target_Spot': 34, 'Tomato_Tomato_Yellow_Leaf_Curl_Virus': 35, 'Tomato_Tomato_mosaic_virus': 36, 'Tomato_healthy': 37}
        
        # Initialize UI
        self.title("Crop Disease Detection")
        self.geometry("600x400")
        self.configure(bg="#f4f4f4")

        self.recent_images = []
        self.diagnosis_history = []
        self.dark_mode = False

        # Create a container for all frames
        container = tk.Frame(self)
        container.pack(fill="both", expand=True)
        
        # Configure grid for centering
        container.grid_rowconfigure(0, weight=1)  
        container.grid_columnconfigure(0, weight=1)

        self.frames = {}
        for F in (HomePage, DiagnosisPage, HistoryPage, FeedbackPage, HelpPage):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame("HomePage")

    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()

    def toggle_dark_mode(self):
        self.dark_mode = not self.dark_mode
        bg_color = "#212121" if self.dark_mode else "#f4f4f4"
        fg_color = "#eeeeee" if self.dark_mode else "#333333"
        
        self.configure(bg=bg_color)
        for frame in self.frames.values():
            frame.configure(bg=bg_color)
            for widget in frame.winfo_children():
                if isinstance(widget, (tk.Label, tk.Button)):
                    widget.config(bg=bg_color, fg=fg_color)

class HomePage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self.configure(bg="#e3f2fd")

        # Title label
        title_label = tk.Label(self, text="Welcome to Crop Disease Diagnosis", font=("Helvetica", 18, "bold"), bg="#e3f2fd", fg="#0d47a1")
        title_label.pack(pady=(40, 20))

        # Navigation buttons
        buttons = [
            ("Go to Diagnosis Page", "DiagnosisPage"),
            ("View Recent Uploads", "HistoryPage"),
            ("Leave Feedback", "FeedbackPage"),
            ("Help", "HelpPage"),
            ("Toggle Dark Mode", None, controller.toggle_dark_mode),
            ("Quit", None, controller.quit)
        ]

        for text, page_name, *command in buttons:
            btn_command = command[0] if command else lambda p=page_name: controller.show_frame(p)
            button = tk.Button(self, text=text, font=("Helvetica", 14), bg="#2196F3", fg="white", cursor="hand2", command=btn_command)
            button.pack(pady=10, padx=20, fill='x')

class DiagnosisPage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self.configure(bg="#ffffff")

        # Title label
        title_label = tk.Label(self, text="Upload Crop Image for Diagnosis", font=("Helvetica", 18, "bold"), bg="#ffffff", fg="#0d47a1")
        title_label.pack(pady=30)

        # Upload button
        upload_button = tk.Button(self, text="Upload Crop Image", font=("Helvetica", 14), bg="#2196F3", fg="white", cursor="hand2", command=self.upload_image)
        upload_button.pack(pady=20)

        # Image label to display uploaded image
        self.image_frame = tk.Frame(self, bg="#e0e0e0", bd=2, relief="sunken")
        self.image_frame.pack(pady=10)
        self.image_label = tk.Label(self.image_frame, bg="#e3f2fd")
        self.image_label.pack()

        # Progress bar
        self.progress = ttk.Progressbar(self, orient="horizontal", length=250, mode="indeterminate")
        self.progress.pack(pady=10)

        # Result label
        self.result_label = tk.Label(self, text="Disease Diagnosis: (None)", font=("Helvetica", 14), bg="#ffffff", fg="#0d47a1")
        self.result_label.pack(pady=20)

        # Save button
        self.save_button = tk.Button(self, text="Save Diagnosis Report", font=("Helvetica", 14), bg="#4CAF50", fg="white", state=tk.DISABLED, cursor="hand2", command=self.save_report)
        self.save_button.pack(pady=10)

        # Filters
        self.create_filter_buttons()

        # Back button
        back_button = tk.Button(self, text="Back to Home", font=("Helvetica", 14), bg="#2196F3", fg="white", cursor="hand2", command=lambda: controller.show_frame("HomePage"))
        back_button.pack(pady=10)

        self.uploaded_image = None

    def create_filter_buttons(self):
        self.filters_frame = tk.Frame(self)
        self.filters_frame.pack(pady=10)

        filters = [
            ("Increase Brightness", self.adjust_brightness), 
            ("Sharpen Image", self.sharpen_image), 
            ("Increase Contrast", self.adjust_contrast), 
            ("Increase Saturation", self.adjust_saturation)
        ]
        for text, cmd in filters:
            btn = tk.Button(self.filters_frame, text=text, font=("Helvetica", 14), bg="#FF5722", fg="white", cursor="hand2", command=cmd)
            btn.pack(side=tk.LEFT, padx=5)

    def upload_image(self):
        file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png")])
        if file_path:
            self.uploaded_image = Image.open(file_path)  # Store the image properly
            self.display_image(self.uploaded_image)
            # Start image processing in a separate thread to avoid blocking the UI
            threading.Thread(target=self.process_image, args=(file_path,)).start()

    def process_image(self, image_path):
        try:
            self.controller.recent_images.append(image_path)
            # Diagnosis in background thread
            prediction = self.diagnose_disease(image_path)
            
            # Update the result on the UI
            self.update_ui(prediction, image_path)
            self.controller.frames["HistoryPage"].refresh_recent_uploads()
        except Exception as e:
            messagebox.showerror("Error", f"An error occurred: {str(e)}")
            
    def update_ui(self, prediction, image_path):
         """Updates UI with the predicted disease and displays the image."""
         # Update the result label with the predicted disease
         self.result_label.config(text=f"Predicted: {prediction}")
         img = Image.open(image_path)
         self.display_image(img)
         self.save_button.config(state=tk.NORMAL)  # Enable save button after diagnosis



    def display_image(self, img):
        img = img.resize((200, 200), Image.Resampling.LANCZOS)  # Resize image
        img_tk = ImageTk.PhotoImage(img)  # Convert to Tkinter-compatible format
        self.image_label.config(image=img_tk)  # Update label with image
        self.image_label.image = img_tk  # Keep reference to prevent garbage collection


    def diagnose_disease(self, image_path):
        try:
            # Load and preprocess the image
            img = image.load_img(image_path, target_size=(150, 150))
            img_array = image.img_to_array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)

            # Predict the disease
            predictions = self.controller.model.predict(img_array)


            # Get the predicted class index
            predicted_class = np.argmax(predictions, axis=1)[0]
            confidence = np.max(predictions) * 100
            if confidence < 50:
                return "Uncertain: please upload a valid crop image"

            # Map the predicted class to its label
            predicted_label = list(self.controller.class_labels.keys())[predicted_class]

            if self.is_non_crop_image(img_array):
                return "Error: Uploaded image does not appear to be a crop image"

            return f"Predicted: {predicted_label} ({confidence:.2f}%)"
        except Exception as e:
            return f"Error: {str(e)}"

    def is_non_crop_image(self,img_array):
        """
        Simple check: If the image does not resemble a crop, return True.
        This function can be replaced with a binary classification model.
        """

        avg_pixel_value = np.mean(img_array)
        if avg_pixel_value > 0.8 or avg_pixel_value < 0.2:
            #Too bright or too dark images might not be crops
            return True
        return False

    def save_report(self):
       diagnosis_text = self.result_label.cget("text")

       if not self.controller.recent_images:
           messagebox.showerror("Error", "No image has been diagnosed yet.")
           return

       file_path = filedialog.asksaveasfilename(defaultextension=".pdf", filetypes=[("PDF files", "*.pdf")])

       if file_path:
          try:
              pdf = FPDF()
              pdf.add_page()
              pdf.set_font("Arial", "B", 16)
              pdf.cell(0, 10, "Diagnosis Report", ln=True, align="C")
              pdf.set_font("Arial", size=12)

               # Get image name safely
              image_name = os.path.basename(self.controller.recent_images[-1])
              pdf.cell(0, 10, f"Image: {image_name}", ln=True)
              pdf.cell(0, 10, f"Diagnosis: {diagnosis_text}", ln=True)

               # Save PDF
              pdf.output(file_path)
              messagebox.showinfo("Report Saved", f"Diagnosis report saved successfully to {file_path}.")

          except Exception as e:
               messagebox.showerror("Error", f"Failed to save report: {str(e)}")
            
    def adjust_brightness(self):
        if self.uploaded_image:
            enhancer = ImageEnhance.Brightness(self.uploaded_image)
            self.apply_filter(enhancer, 1.5)

    def adjust_contrast(self):
        if self.uploaded_image:
            enhancer = ImageEnhance.Contrast(self.uploaded_image)
            self.apply_filter(enhancer, 1.5)

    def sharpen_image(self):
        if self.uploaded_image:
            self.apply_filter(None, filter=ImageFilter.SHARPEN)

    def adjust_saturation(self):
        if self.uploaded_image:
            enhancer = ImageEnhance.Color(self.uploaded_image)
            self.apply_filter(enhancer, 1.5)

    def apply_filter(self, enhancer=None, factor=1.0, filter=None):
        if self.uploaded_image:
            if enhancer:
               self.uploaded_image = enhancer.enhance(factor)  # Update stored image
            elif filter:
                self.uploaded_image = self.uploaded_image.filter(filter)

            self.display_image(self.uploaded_image)  # Refresh UI

class HistoryPage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self.configure(bg="#f3e5f5")

        # Title label
        title_label = tk.Label(self, text="Recent Uploads", font=("Helvetica", 18, "bold"), bg="#f3e5f5", fg="#4a148c")
        title_label.pack(pady=20)

        # Listbox to display recent uploads
        self.recent_uploads = tk.Listbox(self, height=8, bg="#ffffff", fg="#4a148c", font=("Helvetica", 12))
        self.recent_uploads.pack(pady=10, padx=20, fill=tk.BOTH, expand=True)

        # Buttons for actions
        button_frame = tk.Frame(self, bg="#f3e5f5")
        button_frame.pack(pady=10)

        view_button = tk.Button(button_frame, text="View Image", font=("Helvetica", 12), bg="#4A148C", fg="white",
                                command=self.view_image)
        view_button.grid(row=0, column=0, padx=5)

        diagnose_button = tk.Button(button_frame, text="Diagnose Again", font=("Helvetica", 12), bg="#2196F3", fg="white",
                                    command=self.re_diagnose)
        diagnose_button.grid(row=0, column=1, padx=5)

        delete_button = tk.Button(button_frame, text="Delete Upload", font=("Helvetica", 12), bg="#D32F2F", fg="white",
                                  command=self.delete_upload)
        delete_button.grid(row=0, column=2, padx=5)

        export_button = tk.Button(button_frame, text="Export History", font=("Helvetica", 12), bg="#388E3C", fg="white",
                                  command=self.export_history)
        export_button.grid(row=0, column=3, padx=5)

        # Back button
        back_button = tk.Button(self, text="Back to Home", font=("Helvetica", 14), bg="#4A148C", fg="white",
                                cursor="hand2", command=lambda: controller.show_frame("HomePage"))
        back_button.pack(pady=10)

        self.refresh_recent_uploads()

    def refresh_recent_uploads(self):
        """Update the list of recent uploads."""
        self.recent_uploads.delete(0, tk.END)
        for img in self.controller.recent_images:
            self.recent_uploads.insert(tk.END, os.path.basename(img))

    def view_image(self):
        """Opens a preview of the selected image."""
        try:
            selected_index = self.recent_uploads.curselection()[0]
            image_path = self.controller.recent_images[selected_index]

            # Open image in a new window
            top = Toplevel(self)
            top.title("Image Preview")
            img = Image.open(image_path)
            img = img.resize((300, 300), Image.Resampling.LANCZOS)
            img_tk = ImageTk.PhotoImage(img)

            label = tk.Label(top, image=img_tk)
            label.image = img_tk  # Keep reference to prevent garbage collection
            label.pack(padx=10, pady=10)

        except IndexError:
            messagebox.showerror("Error", "Please select an image from the list.")

    def re_diagnose(self):
        """Re-run diagnosis on the selected image."""
        try:
            selected_index = self.recent_uploads.curselection()[0]
            image_path = self.controller.recent_images[selected_index]

            # Switch to DiagnosisPage and start processing again
            self.controller.show_frame("DiagnosisPage")
            self.controller.frames["DiagnosisPage"].process_image(image_path)

        except IndexError:
            messagebox.showerror("Error", "Please select an image from the list.")

    def delete_upload(self):
        """Remove the selected image from the recent uploads list."""
        try:
            selected_index = self.recent_uploads.curselection()[0]
            del self.controller.recent_images[selected_index]
            self.refresh_recent_uploads()

        except IndexError:
            messagebox.showerror("Error", "Please select an image to delete.")

    def export_history(self):
        """Exports the recent upload history to a CSV file."""
        file_path = filedialog.asksaveasfilename(defaultextension=".csv",
                                                 filetypes=[("CSV files", "*.csv")],
                                                 title="Save Upload History")
        if file_path:
            try:
                with open(file_path, "w") as file:
                    file.write("Uploaded Images\n")
                    for img in self.controller.recent_images:
                        file.write(f"{os.path.basename(img)}\n")
                messagebox.showinfo("Export Successful", f"Upload history saved to {file_path}.")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to save history: {str(e)}")

class FeedbackPage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self.configure(bg="#e8f5e9")

        # Title label
        title_label = tk.Label(self, text="User Feedback", font=("Helvetica", 18, "bold"), bg="#e8f5e9", fg="#1b5e20")
        title_label.pack(pady=20)

        # Feedback text area
        self.feedback_text = tk.Text(self, height=8, width=50, font=("Helvetica", 12))
        self.feedback_text.pack(pady=10, padx=20)

        # Submit button
        submit_button = tk.Button(self, text="Submit Feedback", font=("Helvetica", 14), bg="#4CAF50", fg="white",
                                  cursor="hand2", command=self.submit_feedback)
        submit_button.pack(pady=10)

        # Back button
        back_button = tk.Button(self, text="Back to Home", font=("Helvetica", 14), bg="#1b5e20", fg="white",
                                cursor="hand2", command=lambda: controller.show_frame("HomePage"))
        back_button.pack(pady=10)

    def submit_feedback(self):
        """Handles feedback submission and saves it to a file."""
        feedback = self.feedback_text.get("1.0", tk.END).strip()
        if not feedback:
            messagebox.showerror("Error", "Please enter your feedback before submitting.")
            return

        file_path = "feedback.txt"
        try:
            with open(file_path, "a") as file:
                file.write(f"{feedback}\n---\n")
            messagebox.showinfo("Thank You!", "Your feedback has been submitted successfully.")
            self.feedback_text.delete("1.0", tk.END)  # Clear text after submission
        except Exception as e:
            messagebox.showerror("Error", f"Failed to save feedback: {str(e)}")

class HelpPage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self.configure(bg="#e1f5fe")

        # Title label
        title_label = tk.Label(self, text="Help & Instructions", font=("Helvetica", 18, "bold"), bg="#e1f5fe", fg="#01579b")
        title_label.pack(pady=20)

        # Help text
        help_text = (
            "Welcome to the Crop Disease Diagnosis application!\n\n"
            "How to use:\n"
            "1. Click 'Go to Diagnosis Page' on the Home Page.\n"
            "2. Upload an image of a crop leaf for disease detection.\n"
            "3. View the predicted disease result and save the report if needed.\n"
            "4. Check recent uploads in the History Page.\n"
            "5. Use the Feedback Page to submit your suggestions or issues.\n\n"
            "Features:\n"
            "- Dark Mode toggle for better visibility.\n"
            "- Image processing filters to enhance image quality.\n"
            "- Report generation in PDF format.\n\n"
            "For further support, contact: ayushchoubey800@gmail.com"
        )

        help_label = tk.Label(self, text=help_text, font=("Helvetica", 12), bg="#e1f5fe", fg="#01579b", justify="left", wraplength=500)
        help_label.pack(pady=10, padx=20)

        # Back button
        back_button = tk.Button(self, text="Back to Home", font=("Helvetica", 14), bg="#0277bd", fg="white",
                                cursor="hand2", command=lambda: controller.show_frame("HomePage"))
        back_button.pack(pady=10)

if __name__ == "__main__":
    app = App()
    app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog, messagebox, Toplevel
from PIL import Image, ImageTk, ImageEnhance, ImageFilter
import time
import threading
import os
from fpdf import FPDF
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import sys
import logging

logging

class App(tk.Tk):
    def __init__(self):
        super().__init__()

        # Load the model
        self.model = load_model("my_model.keras")

        # Define class labels - make sure these match your model's output
        self.class_labels = {'Apple_Apple_scab': 0, 'Apple_Black_rot': 1, 'Apple_Cedar_apple_rust': 2, 'Apple_healthy': 3, 'Blueberry_healthy': 4, 'Cherry_Powdery_mildew': 5, 'Cherry_healthy': 6, 'Corn_Cercospora_leaf_spot_Gray_leaf_spot': 7, 'Corn_Common_rust': 8, 'Corn_Northern_Leaf_Blight': 9, 'Corn_healthy': 10, 'Grape_Black_rot': 11, 'Grape_Esca_Black_Measles': 12, 'Grape_Leaf_blight': 13, 'Grape_healthy': 14, 'Orange_Haunglongbing_Citrus_greening': 15, 'Peach_Bacterial_spot': 16, 'Peach_healthy': 17, 'Pepper_bell_Bacterial_spot': 18, 'Pepper_bell_healthy': 19, 'Potato_Early_blight': 20, 'Potato_Late_blight': 21, 'Potato_healthy': 22, 'Raspberry_healthy': 23, 'Soybean_healthy': 24, 'Squash_Powdery_mildew': 25, 'Strawberry_Leaf_scorch': 26, 'Strawberry_healthy': 27, 'Tomato_Bacterial_spot': 28, 'Tomato_Early_blight': 29, 'Tomato_Late_blight': 30, 'Tomato_Leaf_Mold': 31, 'Tomato_Septoria_leaf_spot': 32, 'Tomato_Spider_mites _Two_spotted_spider_mite': 33, 'Tomato_Target_Spot': 34, 'Tomato_Tomato_Yellow_Leaf_Curl_Virus': 35, 'Tomato_Tomato_mosaic_virus': 36, 'Tomato_healthy': 37}
        
        # Initialize UI
        self.title("Crop Disease Detection")
        self.geometry("700x500")
        self.configure(bg="#F5F5F5")
        
        self.recent_images = []
        self.diagnosis_history = []
        self.dark_mode = False

        # Create a style for modern buttons
        self.style = ttk.Style()
        self.style.configure('TButton',
                             background='#1E88E5',  # Default button color
                             foreground='#212121',
                             padding=(15, 10),
                             font=('Helvetica', 14),
                             borderwidth=0)
        self.style.map('TButton',
                       background=[('active', '#1976D2')],  # Color when clicked
                       relief=[('active', 'sunken')])  # Slight sunken effect when clicked

        # Style for hover effect
        self.style.configure('Hover.TButton',
                             background='#64B5F6')  # Hover color


        self.frames = {}
        # Create a container for all frames
        container = tk.Frame(self, padx=20, pady=20)
        container.pack(fill="both", expand=True)
        
        # Configure grid for centering
        container.grid_rowconfigure(0, weight=1)  
        container.grid_columnconfigure(0, weight=1)

        self.icon_images = {}
        self.load_icons()

        self.help_page = HelpPage(parent=container, controller=self, icon_images=self.icon_images) # Pass icon_images directly
        self.frames["HelpPage"] = self.help_page
        self.help_page.grid(row=0, column=0, sticky="nsew")
        
        for F in (HomePage, DiagnosisPage, HistoryPage, FeedbackPage, HelpPage):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
            frame.grid(row=0, column=0, sticky="nsew")


        self.show_frame("HomePage")
        

    def get_resource_path(self, relative_path):
        try:
            base_path = sys._MEIPASS
        except Exception:
            base_path = os.path.abspath(".")
        return os.path.join(base_path, relative_path)


        # Load icons FIRST (with error checking)
    def load_icon(self, filename, attribute_name):
        try:
            icon_path = self.get_resource_path(os.path.join("icons", filename))
            print(f"Attempting to load icon: {icon_path}")
            if not os.path.exists(icon_path):
                print(f"Error: Icon file not found: {icon_path}")
                raise FileNotFoundError(f"Icon file not found: {icon_path}")
            img_pil = Image.open(icon_path).resize((20, 20))
            img = ImageTk.PhotoImage(img_pil)
            setattr(self, attribute_name, img)
            self.icon_images[attribute_name] = img
            print(f"Icon '{filename}' loaded successfully. self.icon_images is now: {self.icon_images.keys()}")
        except FileNotFoundError as e:
            print(f"Error loading icon '{filename}': {e}")
            setattr(self, attribute_name, None)
        except Exception as e:
            print(f"An unexpected error occurred loading icon '{filename}': {e}")
            setattr(self, attribute_name, None)


    def load_icons(self):
        icons_dir = self.get_resource_path("icons")
        if not os.path.exists(icons_dir):
            os.makedirs(icons_dir, exist_ok=True)
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234500.png", "home_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234602.png", "upload_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234709.png", "history_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234802.png", "feedback_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235038.png", "help_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235215.png", "dark_mode_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235257.png", "quit_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235406.png", "save_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235500.png", "view_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235616.png", "diagnose_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235700.png", "delete_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235742.png", "export_icon")
        self.load_icon(r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235844.png", "submit_icon")
            

    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()

    def toggle_dark_mode(self):
        self.dark_mode = not self.dark_mode
        bg_color = "#212121" if self.dark_mode else "#F5F5F5"
        fg_color = "#EEEEEE" if self.dark_mode else "#212121"

        self.configure(bg=bg_color)
        for frame in self.frames.values():
            frame.configure(bg=bg_color)
            for widget in frame.winfo_children():
                if isinstance(widget, tk.Frame):
                    widget.config(background=bg_color)
                elif isinstance(widget, ttk.Progressbar):
                    pass
                elif isinstance(widget, tk.PhotoImage):
                    pass
        #Modify ttk style.
        if self.dark_mode:
            self.style.configure('TButton', background = "#64B5F6", foreground = "#212121")
            self.style.configure('Hover.TButton', background = "#90CAF9")
        else:
            self.style.configure('TButton', background = "#1E88E5", foreground = "#212121")
            self.style.configure('Hover.TButton', background = "#64B5F6")

            
    def quit(self):
        self.destroy()

class HomePage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.configure(bg="#F5F5F5")

        # Title label
        title_label = tk.Label(self, text="Welcome to Crop Disease Diagnosis", font=("Helvetica", 18, "bold"), bg="#F5F5F5", fg="#1E88E5")
        title_label.grid(row=0, column=0, columnspan=2, pady=(20, 10))

        # Navigation buttons
        buttons = [
            ("Go to Diagnosis Page", "DiagnosisPage", controller.upload_icon if hasattr(controller, 'upload_icon') else None, lambda p="DiagnosisPage": controller.show_frame(p)),
            ("View Recent Uploads", "HistoryPage", controller.history_icon if hasattr(controller, 'history_icon') else None, lambda p="HistoryPage": controller.show_frame(p)),
            ("Leave Feedback", "FeedbackPage", controller.feedback_icon if hasattr(controller, 'feedback_icon') else None, lambda p="FeedbackPage": controller.show_frame(p)),
            ("Help", "HelpPage", controller.help_icon if hasattr(controller, 'help_icon') else None, lambda p="HelpPage": controller.show_frame(p)),
            ("Toggle Dark Mode", None, controller.dark_mode_icon if hasattr(controller, 'dark_mode_icon') else None, controller.toggle_dark_mode),
            ("Quit", None, controller.quit_icon if hasattr(controller, 'quit_icon') else None, controller.quit)
        ]

        for i, (text, page_name, *command) in enumerate(buttons):
            btn_command = command[-1] if command else None
            print(f"Button '{text}' command: {btn_command}")
            if btn_command:
                if len(command) > 1 and isinstance(command[0], tk.PhotoImage):
                    button = ttk.Button(self, text=text, image=command[0], compound=tk.LEFT, command=btn_command)
                    button.image_ref = command[0]
                    if button["image"] is None:
                        print(f"Button '{text}' image is None.")
                else:
                    button = ttk.Button(self, text=text, command=btn_command)
            else:
                button = ttk.Button(self, text=text)

            button.grid(row=i + 1, column=0, columnspan=2, pady=5, sticky="ew")

            
class DiagnosisPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.configure(bg="#F5F5F5")

        # Title label
        title_label = tk.Label(self, text="Upload Crop Image for Diagnosis", font=("Helvetica", 18, "bold"), bg="#F5F5F5", fg="#1E88E5")
        title_label.grid(row=0, column=0, columnspan=2, pady=(20, 10))

        # Upload button
        self.upload_icon = controller.upload_icon if hasattr(controller, 'upload_icon') and controller.upload_icon else None
        upload_button = ttk.Button(
            self,
            text="Upload Crop Image",
            image=self.upload_icon,
            compound=tk.LEFT,
            command=self.upload_image,
            style='TButton'
        )
        if self.upload_icon:
            upload_button.image = self.upload_icon # Keep a reference!
        upload_button.grid(row=1, column=0, columnspan=2, pady=10)
        upload_button.bind("<Enter>", lambda event, b=upload_button: b.config(style='Hover.TButton'))
        upload_button.bind("<Leave>", lambda event, b=upload_button: b.config(style='TButton'))

        # Image label to display uploaded image
        self.image_frame = tk.Frame(self, bg="#E0E0E0", bd=2, relief="sunken", padx=10, pady=10)
        self.image_frame.grid(row=2, column=0, columnspan=2, pady=10)
        self.image_label = tk.Label(self.image_frame, bg="#F5F5F5")
        self.image_label.pack()

        # Progress bar
        self.progress = ttk.Progressbar(self, orient="horizontal", length=250, mode="indeterminate")
        self.progress.grid(row=3, column=0, columnspan=2, pady=10)
        self.progress.grid_remove() # Initially hide the progress bar

        # Result label
        self.result_label = tk.Label(self, text="Disease Diagnosis: (None)", font=("Helvetica", 14), bg="#F5F5F5", fg="#1E88E5")
        self.result_label.grid(row=4, column=0, columnspan=2, pady=10)

        # Save button
        self.save_icon = controller.save_icon if hasattr(controller, 'save_icon') and controller.save_icon else None
        self.save_button = ttk.Button(
            self,
            text="Save Diagnosis Report",
            image=self.save_icon,
            compound=tk.LEFT,
            state=tk.DISABLED,
            command=self.save_report,
            style='TButton'
        )
        if self.save_icon:
            self.save_button.image = self.save_icon # Keep a reference!
        self.save_button.grid(row=5, column=0, columnspan=2, pady=10)
        self.save_button.bind("<Enter>", lambda event, b=self.save_button: b.config(style='Hover.TButton'))
        self.save_button.bind("<Leave>", lambda event, b=self.save_button: b.config(style='TButton'))

        # Filters
        self.create_filter_buttons()

        # Back button
        back_button = ttk.Button(self, text="Back to Home", command=lambda: controller.show_frame("HomePage"), style='TButton')
        back_button.grid(row=8, column=0, columnspan=2, pady=10) # Moved down below filters
        back_button.bind("<Enter>", lambda event, b=back_button: b.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda event, b=back_button: b.config(style='TButton'))


        self.uploaded_image = None

    def create_filter_buttons(self):
        self.filters_frame = tk.Frame(self, bg="#F5F5F5") # Ensure background matches
        self.filters_frame.grid(row=7, column=0, columnspan=2, pady=10)

        filters = [
            ("Increase Brightness", self.adjust_brightness),
            ("Sharpen Image", self.sharpen_image),
            ("Increase Contrast", self.adjust_contrast),
            ("Increase Saturation", self.adjust_saturation)
        ]
        for i, (text, cmd) in enumerate(filters):
            btn = ttk.Button(self.filters_frame, text=text, command=cmd, style='TButton')
            btn.grid(row=0, column=i, padx=5) # Use grid
            btn.bind("<Enter>", lambda event, b=btn: b.config(style='Hover.TButton'))
            btn.bind("<Leave>", lambda event, b=btn: b.config(style='TButton'))

    def upload_image(self):
        file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png")])
        if file_path:
            try:
                self.uploaded_image = Image.open(file_path)  # Store the PIL
                self.display_image(self.uploaded_image)
                # Start image processing in a separate thread to avoid blocking the UI
                self.progress.grid()
                self.progress.start()
                threading.Thread(target=self.process_image, args=(file_path,)).start()
            except Exception as e:
                messagebox.showerror("Error", f"Failed to open image: {e}")
                logging.error(f"Failed to open image: {e}")

    def process_image(self, image_path):
        try:
            self.controller.recent_images.append(image_path)
            # Diagnosis in background thread
            prediction = self.diagnose_disease(image_path)

            # Update the result on the UI (must be done on the main thread)
            self.after(0, self.update_ui, prediction, image_path)
            self.controller.frames["HistoryPage"].refresh_recent_uploads()
        except Exception as e:
            messagebox.showerror("Error", f"An error occurred during processing: {str(e)}")
            logging.error(f"Error during processing: {str(e)}")
        finally:
            self.after(0, self.progress.stop)
            self.after(0, self.progress.grid_remove)

    def update_ui(self, prediction, image_path):
        """Updates UI with the predicted disease and displays the image."""
        # Update the result label with the predicted disease
        self.result_label.config(text=f"Predicted: {prediction}")
        try:
            img = Image.open(image_path)
            self.display_image(img)
            self.save_button.config(state=tk.NORMAL)  # Enable save button after diagnosis
        except Exception as e:
            messagebox.showerror("Error", f"Failed to display image: {e}")
            logging.error(f"Failed to display image: {e}")


    def display_image(self, img):
        try:
            img = img.resize((200, 200), Image.Resampling.LANCZOS)  # Resize image
            img_tk = ImageTk.PhotoImage(img)  # Convert to Tkinter-compatible format
            self.image_label.config(image=img_tk)  # Update label with image
            self.image_label.image = img_tk  # Keep reference to prevent garbage collection
        except Exception as e:
            messagebox.showerror("Error", f"Failed to display image in UI: {e}")
            logging.error(f"Failed to display image in UI: {e}")


    def diagnose_disease(self, image_path):
        if self.controller.model is None:
            return "Error: Model not loaded."
        try:
            # Load and preprocess the image
            img = image.load_img(image_path, target_size=(150, 150))
            img_array = image.img_to_array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)

            # Predict the disease
            predictions = self.controller.model.predict(img_array)

            # Get the predicted class index
            predicted_class = np.argmax(predictions, axis=1)[0]
            confidence = np.max(predictions) * 100
            if confidence < 50:
                return "Uncertain: please upload a valid crop image"

            # Map the predicted class to its label
            predicted_label = list(self.controller.class_labels.keys())[predicted_class]

            if self.is_non_crop_image(img_array):
                return "Error: Uploaded image does not appear to be a crop image"

            return f"Predicted: {predicted_label} ({confidence:.2f}%)"
        except Exception as e:
            return f"Error during diagnosis: {str(e)}"

    def is_non_crop_image(self,img_array):
        """
        Simple check: If the image does not resemble a crop, return True.
        This function can be replaced with a binary classification model.
        """
        avg_pixel_value = np.mean(img_array)
        if avg_pixel_value > 0.8 or avg_pixel_value < 0.2:
            #Too bright or too dark images might not be crops
            return True
        return False

    def save_report(self):
        diagnosis_text = self.result_label.cget("text")

        if not self.controller.recent_images:
            messagebox.showerror("Error", "No image has been diagnosed yet.")
            return

        file_path = filedialog.asksaveasfilename(defaultextension=".pdf", filetypes=[("PDF files", "*.pdf")])

        if file_path:
            try:
                pdf = FPDF()
                pdf.add_page()
                pdf.set_font("Arial", "B", 16)
                pdf.cell(0, 10, "Diagnosis Report", ln=True, align="C")
                pdf.set_font("Arial", size=12)

                # Get image name safely
                image_name = os.path.basename(self.controller.recent_images[-1])
                pdf.cell(0, 10, f"Image: {image_name}", ln=True)
                pdf.cell(0, 10, f"Diagnosis: {diagnosis_text}", ln=True)

                # Save PDF
                pdf.output(file_path)
                messagebox.showinfo("Report Saved", f"Diagnosis report saved successfully to {file_path}.")

            except Exception as e:
                messagebox.showerror("Error", f"Failed to save report: {str(e)}")

    def adjust_brightness(self):
        if self.uploaded_image:
            enhancer = ImageEnhance.Brightness(self.uploaded_image)
            self.apply_filter(enhancer, 1.5)

    def adjust_contrast(self):
        if self.uploaded_image:
            enhancer = ImageEnhance.Contrast(self.uploaded_image)
            self.apply_filter(enhancer, 1.5)

    def sharpen_image(self):
        if self.uploaded_image:
            self.apply_filter(None, filter=ImageFilter.SHARPEN)

    def adjust_saturation(self):
        if self.uploaded_image:
            enhancer = ImageEnhance.Color(self.uploaded_image)
            self.apply_filter(enhancer, 1.5)

    def apply_filter(self, enhancer=None, factor=1.0, filter=None):
        if self.uploaded_image:
            try:
                if enhancer:
                    self.uploaded_image = enhancer.enhance(factor)  # Update stored image
                elif filter:
                    self.uploaded_image = self.uploaded_image.filter(filter)

                self.display_image(self.uploaded_image)  # Refresh UI
            except Exception as e:
                messagebox.showerror("Error", f"Error applying filter: {e}")
                logging.error(f"Error applying filter: {e}")

class HistoryPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.configure(bg="#E0E0E0")

        # Title label
        title_label = tk.Label(self, text="Recent Uploads", font=("Helvetica", 18, "bold"), bg="#E0E0E0", fg="#1E88E5")
        title_label.grid(row=0, column=0, columnspan=4, pady=(20, 10))

        # Listbox to display recent uploads
        self.recent_uploads = tk.Listbox(self, height=8, bg="#F5F5F5", fg="#212121", font=("Helvetica", 12))
        self.recent_uploads.grid(row=1, column=0, columnspan=4, pady=10, padx=20, sticky="nsew")


        # Buttons for actions
        button_frame = tk.Frame(self, bg="#E0E0E0", padx=10, pady=10)
        button_frame.grid(row=2, column=0, columnspan=4, pady=10)

        self.view_icon = controller.view_icon if hasattr(controller, 'view_icon') and controller.view_icon else None
        view_button = ttk.Button(
            button_frame,
            text="View Image",
            image=self.view_icon,
            compound=tk.LEFT,
            command=self.view_image,
            style='TButton'
        )
        if self.view_icon:
            view_button.image = self.view_icon # Keep a reference!
        view_button.grid(row=0, column=0, padx=5, pady=5)
        view_button.bind("<Enter>", lambda event, b=view_button: b.config(style='Hover.TButton'))
        view_button.bind("<Leave>", lambda event, b=view_button: b.config(style='TButton'))

        self.diagnose_icon = controller.diagnose_icon if hasattr(controller, 'diagnose_icon') and controller.diagnose_icon else None
        diagnose_button = ttk.Button(
            button_frame,
            text="Diagnose Again",
            image=self.diagnose_icon,
            compound=tk.LEFT,
            command=self.re_diagnose,
            style='TButton'
        )
        if self.diagnose_icon:
            diagnose_button.image = self.diagnose_icon # Keep a reference!
        diagnose_button.grid(row=0, column=1, padx=5, pady=5)
        diagnose_button.bind("<Enter>", lambda event, b=diagnose_button: b.config(style='Hover.TButton'))
        diagnose_button.bind("<Leave>", lambda event, b=diagnose_button: b.config(style='TButton'))

        self.delete_icon = controller.delete_icon if hasattr(controller, 'delete_icon') and controller.delete_icon else None
        delete_button = ttk.Button(
            button_frame,
            text="Delete Upload",
            image=self.delete_icon,
            compound=tk.LEFT,
            command=self.delete_upload,
            style='TButton'
        )
        if self.delete_icon:
            delete_button.image = self.delete_icon # Keep a reference!
        delete_button.grid(row=0, column=2, padx=5, pady=5)
        delete_button.bind("<Enter>", lambda event, b=delete_button: b.config(style='Hover.TButton'))
        delete_button.bind("<Leave>", lambda event, b=delete_button: b.config(style='TButton'))

        self.export_icon = controller.export_icon if hasattr(controller, 'export_icon') and controller.export_icon else None
        export_button = ttk.Button(
            button_frame,
            text="Export History",
            image=self.export_icon,
            compound=tk.LEFT,
            command=self.export_history,
            style='TButton'
        )
        if self.export_icon:
            export_button.image = self.export_icon # Keep a reference!
        export_button.grid(row=0, column=3, padx=5, pady=5)
        export_button.bind("<Enter>", lambda event, b=export_button: b.config(style='Hover.TButton'))
        export_button.bind("<Leave>", lambda event, b=export_button: b.config(style='TButton'))

        # Back button
        back_button = ttk.Button(self, text="Back to Home", command=lambda: controller.show_frame("HomePage"), style='TButton')
        back_button.grid(row=3, column=0, columnspan=4, pady=10)
        back_button.bind("<Enter>", lambda event, b=back_button: b.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda event, b=back_button: b.config(style='TButton'))


        self.refresh_recent_uploads()

    def refresh_recent_uploads(self):
        """Update the list of recent uploads."""
        self.recent_uploads.delete(0, tk.END)
        for img in self.controller.recent_images:
            self.recent_uploads.insert(tk.END, os.path.basename(img))

    def view_image(self):
        """Opens a preview of the selected image."""
        try:
            selected_index = self.recent_uploads.curselection()[0]
            image_path = self.controller.recent_images[selected_index]

            # Open image in a new window
            top = Toplevel(self)
            top.title("Image Preview")
            try:
                img = Image.open(image_path)
                img = img.resize((300, 300), Image.Resampling.LANCZOS)
                img_tk = ImageTk.PhotoImage(img)

                label = tk.Label(top, image=img_tk)
                label.image = img_tk  # Keep reference
                label.pack(padx=10, pady=10)
            except Exception as e:
                messagebox.showerror("Error", f"Failed to open image: {e}")
                logging.error(f"Failed to open image: {e}")

        except IndexError:
            messagebox.showerror("Error", "Please select an image from the list.")

    def re_diagnose(self):
        """Re-run diagnosis on the selected image."""
        try:
            selected_index = self.recent_uploads.curselection()[0]
            image_path = self.controller.recent_images[selected_index]

            # Switch to DiagnosisPage and start processing again
            self.controller.show_frame("DiagnosisPage")
            self.controller.frames["DiagnosisPage"].process_image(image_path)

        except IndexError:
            messagebox.showerror("Error", "Please select an image from the list.")

    def delete_upload(self):
        """Remove the selected image from the recent uploads list."""
        try:
            selected_index = self.recent_uploads.curselection()[0]
            del self.controller.recent_images[selected_index]
            self.refresh_recent_uploads()

        except IndexError:
            messagebox.showerror("Error", "Please select an image to delete.")

    def export_history(self):
        """Exports the recent upload history to a CSV file."""
        file_path = filedialog.asksaveasfilename(defaultextension=".csv",
                                                 filetypes=[("CSV files", "*.csv")],
                                                 title="Save Upload History")
        if file_path:
            try:
                with open(file_path, "w") as file:
                    file.write("Uploaded Images\n")
                    for img in self.controller.recent_images:
                        file.write(f"{os.path.basename(img)}\n")
                messagebox.showinfo("Export Successful", f"Upload history saved to {file_path}.")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to save history: {str(e)}")

class FeedbackPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.configure(bg="#E0E0E0")

        # Title label
        title_label = tk.Label(self, text="User Feedback", font=("Helvetica", 18, "bold"), bg="#E0E0E0", fg="#1E88E5")
        title_label.grid(row=0, column=0, columnspan=2, pady=(20, 10))

        # Feedback text area
        self.feedback_text = tk.Text(self, height=8, width=50, font=("Helvetica", 12), bg="#F5F5F5", fg="#212121", padx=5, pady=5)
        self.feedback_text.grid(row=1, column=0, columnspan=2, pady=10, padx=20)
        
        # Submit button
        self.submit_icon = controller.submit_icon if hasattr(controller, 'submit_icon') and controller.submit_icon else None
        submit_button = ttk.Button(
            self,
            text="Submit Feedback",
            image=self.submit_icon,
            compound=tk.LEFT,
            command=self.submit_feedback,
            style='TButton'
        )
        if self.submit_icon:
            submit_button.image = self.submit_icon # Keep a reference!
        submit_button.grid(row=2, column=0, columnspan=2, pady=10)
        submit_button.bind("<Enter>", lambda event, b=submit_button: b.config(style='Hover.TButton'))
        submit_button.bind("<Leave>", lambda event, b=submit_button: b.config(style='TButton'))


        # Back button
        back_button = ttk.Button(self, text="Back to Home", command=lambda: controller.show_frame("HomePage"), style='TButton')
        back_button.grid(row=3, column=0, columnspan=2, pady=10)
        back_button.bind("<Enter>", lambda event, b=back_button: b.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda event, b=back_button: b.config(style='TButton'))
        
    def submit_feedback(self):
        """Handles feedback submission and saves it to a file."""
        feedback = self.feedback_text.get("1.0", tk.END).strip()
        if not feedback:
            messagebox.showerror("Error", "Please enter your feedback before submitting.")
            return

        file_path = "feedback.txt"
        try:
            with open(file_path, "a") as file:
                file.write(f"{feedback}\n---\n")
            messagebox.showinfo("Thank You!", "Your feedback has been submitted successfully.")
            self.feedback_text.delete("1.0", tk.END)  # Clear text after submission
        except Exception as e:
            messagebox.showerror("Error", f"Failed to save feedback: {str(e)}")
            logging.error(f"Failed to save feedback: {str(e)}")

class HelpPage(tk.Frame):
    def __init__(self, parent, controller, icon_images):
        super().__init__(parent)
        self.controller = controller
        self.icon_images = icon_images
        self.configure(bg="#E0E0E0")
        self.icon_labels = {}

        # Title label
        title_label = tk.Label(self, text="Help & Instructions", font=("Helvetica", 20, "bold"), bg="#E0E0E0", fg="#1E88E5", pady=10)
        title_label.grid(row=0, column=0, columnspan=2, pady=(20, 10), sticky="ew")

        self.help_sections_data = [
            {
                "title": "Getting Started",
                "icon": "play.png",  # Placeholder icon
                "content": "Welcome to the Crop Disease Diagnosis application! This application helps you identify diseases in crop leaves using image analysis. Follow the steps below to get started."
            },
            {
                "title": "How to Diagnose a Disease",
                "icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235616.png",  # Placeholder icon
                "steps": [
                    "Navigate to the 'Diagnosis Page' by clicking the button on the Home Page.",
                    "Click the 'Upload Crop Image' button to select an image of a crop leaf from your computer.",
                    "The application will process the image and display the predicted disease along with a confidence score.",
                    "You can optionally apply image filters (brightness, contrast, sharpen, saturation) before or after diagnosis to potentially improve results.",
                    "Click 'Save Diagnosis Report' to save the diagnosis and the image information as a PDF file."
                ]
            },
            {
                "title": "Viewing Recent Uploads (History Page)",
                "icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234709.png",  # Placeholder icon
                "content": "The 'Recent Uploads' page allows you to view a history of the images you've uploaded.",
                "actions": [
                    "**View Image:** Select an image from the list and click 'View Image' to see a larger preview.",
                    "**Diagnose Again:** Select an image and click 'Diagnose Again' to re-run the disease detection process on that image.",
                    "**Delete Upload:** Select an image and click 'Delete Upload' to remove it from your recent history.",
                    "**Export History:** Click 'Export History' to save the list of your recent uploads to a CSV file."
                ]
            },
            {
                "title": "Providing Feedback",
                "icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234802.png",  # Placeholder icon
                "content": "We value your feedback! If you have any suggestions, bug reports, or comments, please use the 'Leave Feedback' page. Simply type your message in the text area and click 'Submit Feedback'."
            },
            {
                "title": "Using Dark Mode",
                "icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235215.png",  # Placeholder icon
                "content": "You can toggle between a light and dark color scheme by clicking the 'Toggle Dark Mode' button on the Home Page. Dark mode can be easier on the eyes in low-light conditions."
            },
            {
                "title": "Getting Further Assistance",
                "icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235038.png",  # Placeholder icon
                "content": f"If you encounter any issues or have further questions, please don't hesitate to contact us at: ayushchoubey800@gmail.com"
            }
        ]

        current_row = 1
        for i, section in enumerate(self.help_sections_data):
            # Icon (if available)
            icon_img = self.controller.icon_images.get(section.get("icon").split(".")[0] + "_icon") if section.get("icon") else None
            icon_label = tk.Label(self, image=icon_img, bg="#E0E0E0") if icon_img else tk.Label(self, text="", bg="#E0E0E0")
            icon_label.grid(row=current_row, column=0, padx=(20, 5), pady=10, sticky="nsw")
            self.icon_labels[i] = icon_label

            # Title
            title_label = tk.Label(self, text=section["title"], font=("Helvetica", 14, "bold"), bg="#E0E0E0", fg="#333333", anchor="w")
            title_label.grid(row=current_row, column=1, columnspan=2, padx=(5, 20), pady=10, sticky="new")
            current_row += 1

            # Content
            if "content" in section:
                content_label = tk.Label(self, text=section["content"], font=("Helvetica", 12), bg="#E0E0E0", fg="#555555", justify="left", wraplength=600, padx=20, pady=5)
                content_label.grid(row=current_row, column=0, columnspan=3, sticky="ew")
                current_row += 1

            # Steps
            if "steps" in section:
                for i, step in enumerate(section["steps"]):
                    step_label_num = tk.Label(self, text=f"{i+1}.", font=("Helvetica", 12, "bold"), bg="#E0E0E0", fg="#555555")
                    step_label_num.grid(row=current_row, column=0, padx=(40, 5), sticky="nw")
                    step_label = tk.Label(self, text=step, font=("Helvetica", 12), bg="#E0E0E0", fg="#555555", justify="left", wraplength=580, anchor="w")
                    step_label.grid(row=current_row, column=1, columnspan=2, sticky="ew", padx=(5, 20), pady=2)
                    current_row += 1

            # Actions
            if "actions" in section:
                for action in section["actions"]:
                    action_label_bullet = tk.Label(self, text="•", font=("Helvetica", 12, "bold"), bg="#E0E0E0", fg="#555555")
                    action_label_bullet.grid(row=current_row, column=0, padx=(40, 5), sticky="nw")
                    action_label = tk.Label(self, text=action, font=("Helvetica", 12), bg="#E0E0E0", fg="#555555", justify="left", wraplength=580, anchor="w")
                    action_label.grid(row=current_row, column=1, columnspan=2, sticky="ew", padx=(5, 20), pady=2)
                    current_row += 1

            # Add a small vertical separator between sections
            separator = ttk.Separator(self, orient="horizontal")
            separator.grid(row=current_row, column=0, columnspan=3, pady=15, sticky="ew")
            current_row += 1

        # Back Button at the very end
        back_button = ttk.Button(self, text="Back to Home", command=lambda: controller.show_frame("HomePage"), style='TButton')
        back_button.grid(row=current_row, column=0, columnspan=3, pady=20)
        back_button.bind("<Enter>", lambda event, b=back_button: b.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda event, b=back_button: b.config(style='TButton'))

        
if __name__ == "__main__":
    app = App()
    app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog, messagebox, Toplevel
from PIL import Image, ImageTk, ImageEnhance, ImageFilter, ImageOps
import time
import threading
import os
from fpdf import FPDF
import numpy as np
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import sys
import logging
from datetime import datetime

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s', filename='crop_diagnosis.log')

class App(tk.Tk):
    def __init__(self):
        super().__init__()

        # Initialize models
        self.disease_model = None
        self.crop_classifier = None
        self.class_labels = None

        self.frames = {}
        
        try:
            # Load the model with error handling
            self.load_models()
            # Define class labels - make sure these match your model's output
            self.class_labels = {
                'Apple_Apple_scab': 0, 'Apple_Black_rot': 1, 'Apple_Cedar_apple_rust': 2, 
                'Apple_healthy': 3, 'Blueberry_healthy': 4, 'Cherry_Powdery_mildew': 5, 
                'Cherry_healthy': 6, 'Corn_Cercospora_leaf_spot_Gray_leaf_spot': 7, 
                'Corn_Common_rust': 8, 'Corn_Northern_Leaf_Blight': 9, 'Corn_healthy': 10, 
                'Grape_Black_rot': 11, 'Grape_Esca_Black_Measles': 12, 'Grape_Leaf_blight': 13, 
                'Grape_healthy': 14, 'Orange_Haunglongbing_Citrus_greening': 15, 
                'Peach_Bacterial_spot': 16, 'Peach_healthy': 17, 'Pepper_bell_Bacterial_spot': 18, 
                'Pepper_bell_healthy': 19, 'Potato_Early_blight': 20, 'Potato_Late_blight': 21, 
                'Potato_healthy': 22, 'Raspberry_healthy': 23, 'Soybean_healthy': 24, 
                'Squash_Powdery_mildew': 25, 'Strawberry_Leaf_scorch': 26, 'Strawberry_healthy': 27, 
                'Tomato_Bacterial_spot': 28, 'Tomato_Early_blight': 29, 'Tomato_Late_blight': 30, 
                'Tomato_Leaf_Mold': 31, 'Tomato_Septoria_leaf_spot': 32, 
                'Tomato_Spider_mites _Two_spotted_spider_mite': 33, 'Tomato_Target_Spot': 34, 
                'Tomato_Tomato_Yellow_Leaf_Curl_Virus': 35, 'Tomato_Tomato_mosaic_virus': 36, 
                'Tomato_healthy': 37
            }
        except Exception as e:
            logging.error(f"Failed to load model: {e}")
            messagebox.showerror("Error", "Failed to load the disease detection model. The application may not function properly.")

        # Initialize UI
        self.title("Crop Disease Detection")
        self.geometry("1000x700")
        self.minsize(800, 600)
        
        self.recent_images = []
        self.diagnosis_history = []
        self.dark_mode = False
        self.icon_images = {}

        self.create_styles()
        self.load_icons()

        # Create container for all frames
        container = ttk.Frame(self)
        container.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
        container.grid_rowconfigure(0, weight=1)  
        container.grid_columnconfigure(0, weight=1)

        # Create all frames
        for F in (HomePage, DiagnosisPage, HistoryPage, FeedbackPage, HelpPage):
            page_name = F.__name__
            if page_name == "HelpPage":
                frame = F(parent=container, controller=self, icon_images=self.icon_images)
            else:
                frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
            frame.grid(row=0, column=0, sticky="nsew", padx=5, pady=5)

        self.show_frame("HomePage")

    def create_styles(self):
        """Create modern UI styles"""
        self.style = ttk.Style()
        
        # Light mode styles
        self.style.configure('.', font=('Helvetica', 12), background='#F5F5F5', foreground='#212121')
        
        self.style.configure('TButton',
                           background='#1E88E5',
                           foreground='white',
                           padding=10,
                           borderwidth=0)
        self.style.map('TButton',
                      background=[('active', '#1976D2'), ('disabled', '#BBDEFB')],
                      foreground=[('disabled', '#757575')])
        
        self.style.configure('TFrame', background='#F5F5F5')
        self.style.configure('TLabel', background='#F5F5F5', foreground='#212121')
        self.style.configure('TLabelframe', background='#F5F5F5')
        self.style.configure('TLabelframe.Label', background='#F5F5F5', foreground='#212121')
        self.style.configure('TEntry', fieldbackground='white', foreground='#212121')
        self.style.configure('TCombobox', fieldbackground='white', foreground='#212121')
        self.style.configure('TProgressbar', background='#1E88E5', troughcolor='#E0E0E0')

    def load_models(self):
        """Load ML models with error handling"""
        model_files = {
            "disease_model": "my_model.keras",
            #"crop_classifier": "crop_classifier.keras"  # optional
        }
        
        for model_name, filename in model_files.items():
            try:
                model_path = self.get_resource_path(f"models/{filename}")
                if not os.path.exists(model_path):
                    logging.warning(f"Model file not found: {model_path}")
                    continue
                    
                setattr(self, model_name, load_model(model_path))
                logging.info(f"Loaded {model_name} successfully")
                
            except Exception as e:
                logging.error(f"Failed to load {model_name}: {e}")
                if model_name == "disease_model":
                    raise RuntimeError(f"Critical model failed to load: {e}")

    def get_resource_path(self, relative_path):
        """Get absolute path to resource, works for dev and for PyInstaller"""
        try:
            # PyInstaller creates a temp folder and stores path in _MEIPASS
            base_path = sys._MEIPASS
        except Exception:
            base_path = os.path.abspath(".")
        
        full_path = os.path.join(base_path, relative_path)
        logging.info(f"Resource path resolved: {full_path}")
        return full_path

    def load_icons(self):
        """Load all icons with error handling"""
        icon_names = {
            "home_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234500.png",
            "upload_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234602.png",
            "history_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234709.png",
            "feedback_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 234802.png",
            "help_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235038.png",
            "dark_mode_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235215.png",
            "quit_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235257.png",
            "save_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235406.png",
            "view_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235500.png",
            "diagnose_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235616.png",
            "delete_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235700.png",
            "export_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235742.png",
            "submit_icon": r"C:\Users\ayush\OneDrive\Desktop\project 1\icon image\Screenshot 2025-04-05 235844.png"
        }

        self.icon_images = {}
        for name, filename in icon_names.items():
            try:
                # Try multiple possible locations
                possible_paths = [
                    os.path.join("icons", filename),
                    self.get_resource_path(f"icons/{filename}"),
                    os.path.join(os.path.dirname(__file__), "icons", filename)
                ]
                
                for path in possible_paths:
                    if os.path.exists(path):
                        img = Image.open(path).resize((24, 24))
                        self.icon_images[name] = ImageTk.PhotoImage(img)
                        break
                else:
                    logging.warning(f"Icon not found: {filename}")
                    self.icon_images[name] = None
                    
            except Exception as e:
                logging.error(f"Error loading icon {filename}: {e}")
                self.icon_images[name] = None


    def toggle_dark_mode(self):
        """Toggle between light and dark mode"""
        self.dark_mode = not self.dark_mode
        
        if self.dark_mode:
            # Dark mode colors
            bg_color = "#212121"
            fg_color = "#EEEEEE"
            button_bg = "#424242"
            button_fg = "#FFFFFF"
            hover_bg = "#616161"
            disabled_bg = "#757575"
            entry_bg = "#424242"
        else:
            # Light mode colors
            bg_color = "#F5F5F5"
            fg_color = "#212121"
            button_bg = "#1E88E5"
            button_fg = "#FFFFFF"
            hover_bg = "#1976D2"
            disabled_bg = "#BBDEFB"
            entry_bg = "white"

        
        # Update style
        self.style.configure('.', background=bg_color, foreground=fg_color, fieldbackground=entry_bg)
        self.style.configure('TButton',
                           background=button_bg,
                           foreground=button_fg)
        self.style.map('TButton',
                      background=[('active', hover_bg), ('disabled', disabled_bg)])
        
        self.style.configure('TFrame', background=bg_color)
        self.style.configure('TLabel', background=bg_color, foreground=fg_color)
        self.style.configure('TLabelframe', background=bg_color)
        self.style.configure('TLabelframe.Label', background=bg_color, foreground=fg_color)
        self.style.configure('TEntry', fieldbackground=entry_bg, foreground=fg_color)
        self.style.configure('TCombobox', fieldbackground=entry_bg, foreground=fg_color)
        
        # Update all widgets
        for frame in self.frames.values():
            if isinstance(frame, tk.Frame):
                frame.configure(bg=bg_color)  # Only for regular tk.Frame
            self.update_widget_colors(frame, bg_color, fg_color)

    def update_widget_colors(self, widget, bg_color, fg_color):
        """Recursively update widget colors"""
        if isinstance(widget, tk.Frame):
            widget.configure(bg=bg_color)
        elif isinstance(widget, tk.Label):
            widget.configure(bg=bg_color, fg=fg_color)
        elif isinstance(widget, (tk.Text, tk.Listbox)):
            widget.configure(bg="#424242" if self.dark_mode else "#FFFFFF", 
                        fg=fg_color, 
                        insertbackground=fg_color)
    
        # Recursively update children
        for child in widget.winfo_children():
            self.update_widget_colors(child, bg_color, fg_color)

    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()
        if hasattr(frame, "on_show"):
            frame.on_show()

    def quit(self):
        self.destroy()

    # Improved non-crop detection methods
    def is_non_crop_image(self, img_path):
        """
        Detect non-crop images using either:
        1. ML classifier (if available) - recommended
        2. Enhanced heuristic approach - fallback
        """
        if self.crop_classifier:
            return self._ml_non_crop_detection(img_path)
        else:
            return self._heuristic_non_crop_detection(img_path)

    def _ml_non_crop_detection(self, img_path):
        """Use trained classifier to detect non-crop images"""
        try:
            # Load and preprocess image
            img = image.load_img(img_path, target_size=(224, 224))
            img_array = image.img_to_array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)
            
            # Predict (assuming model outputs 0 for non-crop, 1 for crop)
            prediction = self.crop_classifier.predict(img_array)
            return prediction[0][0] < 0.5  # Threshold can be adjusted
            
        except Exception as e:
            logging.error(f"ML non-crop detection failed: {e}")
            return False  # Default to assuming it's a crop

    def _heuristic_non_crop_detection(self, img_path):
        """Enhanced heuristic approach for non-crop detection"""
        try:
            # Open image and convert to numpy array
            img = Image.open(img_path)
            img_array = np.array(img)
            
            # Convert to HSV for color analysis
            hsv = cv2.cvtColor(img_array, cv2.COLOR_RGB2HSV)
            
            # 1. Check green color dominance (crop leaves are mostly green)
            green_pixels = np.sum((hsv[:,:,0] > 30) & (hsv[:,:,0] < 90))
            green_ratio = green_pixels / (img.size[0] * img.size[1])
            
            # 2. Check texture (crop leaves have more texture)
            gray = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)
            laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()
            
            # 3. Check brightness/contrast
            avg_brightness = np.mean(gray)
            contrast = gray.std()
            
            # Decision thresholds (tuned empirically)
            if (green_ratio < 0.3 or      # Not enough green
                laplacian_var < 100 or    # Too smooth
                avg_brightness < 30 or    # Too dark
                avg_brightness > 220 or   # Too bright
                contrast < 40):           # Low contrast
                return True
                
            return False
            
        except Exception as e:
            logging.error(f"Heuristic non-crop detection failed: {e}")
            return False  # Default to assuming it's a crop


class HomePage(ttk.Frame):
    def __init__(self, parent, controller):
        ttk.Frame.__init__(self, parent)
        self.controller = controller

        # Configure grid
        self.grid_rowconfigure(0, weight=1)
        self.grid_columnconfigure(0, weight=1)
        
        # Main container
        main_frame = ttk.Frame(self)
        main_frame.grid(row=0, column=0, sticky="nsew", padx=20, pady=20)
        main_frame.grid_rowconfigure(1, weight=1)
        main_frame.grid_columnconfigure(0, weight=1)

        # Title label
        title_label = ttk.Label(
            main_frame, 
            text="Welcome to Crop Disease Diagnosis", 
            font=("Helvetica", 20, "bold")
        )
        title_label.grid(row=0, column=0, pady=(0, 30), padx=10)

        # Button frame
        button_frame = ttk.Frame(main_frame)
        button_frame.grid(row=1, column=0, sticky="nsew", padx=10, pady=10)
        
        # Navigation buttons
        buttons = [
            ("Diagnose Crop", "upload_icon", "DiagnosisPage"),
            ("View History", "history_icon", "HistoryPage"),
            ("Provide Feedback", "feedback_icon", "FeedbackPage"),
            ("Help & Guide", "help_icon", "HelpPage"),
            ("Toggle Dark Mode", "dark_mode_icon", None, lambda: controller.toggle_dark_mode()),
            ("Exit Application", "quit_icon", None, lambda: controller.quit())
        ]

        for i, (text, icon_name, page_name, *command) in enumerate(buttons):
            btn_frame = ttk.Frame(button_frame)
            btn_frame.grid(row=i, column=0, pady=5, padx=5, sticky="ew")
            btn_frame.grid_columnconfigure(0, weight=1)
            
            btn_command = command[0] if command else (lambda p=page_name: controller.show_frame(p))
            icon = controller.icon_images.get(icon_name)
            
            btn = ttk.Button(
                btn_frame,
                text=text,
                image=icon,
                compound=tk.LEFT,
                command=btn_command,
                style='TButton'
            )
            btn.grid(sticky="ew", padx=5, pady=5)
            
            # Add hover effect
            btn.bind("<Enter>", lambda e, b=btn: b.config(style='Hover.TButton'))
            btn.bind("<Leave>", lambda e, b=btn: b.config(style='TButton'))

        

class DiagnosisPage(ttk.Frame):
    def __init__(self, parent, controller):
        ttk.Frame.__init__(self, parent)
        self.controller = controller
        self.uploaded_image = None
        self.current_image_path = None
        self.filtered_image = None

        # Configure grid weights
        self.grid_rowconfigure(1, weight=1)
        self.grid_columnconfigure(0, weight=1)

        # Title label
        title_label = ttk.Label(self, text="Upload Crop Image for Diagnosis", 
                             font=("Helvetica", 18, "bold"))
        title_label.grid(row=0, column=0, columnspan=2, pady=(0, 20), padx=10)

        # Upload button
        upload_button = ttk.Button(
            self,
            text="Upload Crop Image",
            image=controller.icon_images.get("upload_icon"),
            compound=tk.LEFT,
            command=self.upload_image,
            style='TButton'
        )
        upload_button.grid(row=1, column=0, columnspan=2, pady=10, padx=10, sticky="ew")
        upload_button.bind("<Enter>", lambda e: upload_button.config(style='Hover.TButton'))
        upload_button.bind("<Leave>", lambda e: upload_button.config(style='TButton'))

        # Image display frame
        self.image_frame = ttk.LabelFrame(self, text="Image Preview")
        self.image_frame.grid(row=2, column=0, columnspan=2, pady=10, padx=10, sticky="nsew")
        self.image_frame.grid_rowconfigure(0, weight=1)
        self.image_frame.grid_columnconfigure(0, weight=1)
        
        self.image_label = ttk.Label(self.image_frame)
        self.image_label.grid(row=0, column=0, sticky="nsew", padx=5, pady=5)

        # Progress bar
        self.progress = ttk.Progressbar(self, orient="horizontal", length=300, mode="determinate")
        self.progress.grid(row=3, column=0, columnspan=2, pady=10, padx=10)
        self.progress.grid_remove()

        self.status_label = ttk.Label(self, text="Ready to upload image", font=("Helvetica", 10))
        self.status_label.grid(row=4, column=0, columnspan=2, pady=5, padx=10)

        # Result label
        result_frame = ttk.LabelFrame(self, text="Diagnosis Results")
        result_frame.grid(row=5, column=0, columnspan=2, pady=10, padx=10, sticky="ew")
        
        self.result_label = ttk.Label(result_frame, text="No diagnosis available", 
                                   font=("Helvetica", 12), wraplength=600)
        self.result_label.pack(padx=10, pady=10)

        # Action buttons
        button_frame = ttk.Frame(self)
        button_frame.grid(row=6, column=0, columnspan=2, pady=10, padx=10, sticky="ew")

        # Save button
        self.save_button = ttk.Button(
            button_frame,
            text="Save Report",
            image=controller.icon_images.get("save_icon"),
            compound=tk.LEFT,
            state=tk.DISABLED,
            command=self.save_report,
            style='TButton'
        )
        self.save_button.pack(side=tk.LEFT, padx=5, pady=5, expand=True, fill=tk.X)
        self.save_button.bind("<Enter>", lambda e: self.save_button.config(style='Hover.TButton'))
        self.save_button.bind("<Leave>", lambda e: self.save_button.config(style='TButton'))

        back_button = ttk.Button(
            button_frame,
            text="Back to Home",
            command=lambda: controller.show_frame("HomePage"),
            style='TButton'
        )
        back_button.pack(side=tk.LEFT, padx=5, pady=5, expand=True, fill=tk.X)
        back_button.bind("<Enter>", lambda e: back_button.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda e: back_button.config(style='TButton'))

        self.create_filter_controls()

    def create_filter_controls(self):
        """Create image filter controls"""
        filter_frame = ttk.LabelFrame(self, text="Image Adjustments")
        filter_frame.grid(row=7, column=0, columnspan=2, pady=10, padx=10, sticky="ew")
        
        # Brightness control
        ttk.Label(filter_frame, text="Brightness:").grid(row=0, column=0, padx=5, pady=5, sticky="e")
        self.brightness_scale = ttk.Scale(
            filter_frame,
            from_=0.1,
            to=2.0,
            value=1.0,
            command=lambda v: self.adjust_brightness(float(v))
        )
        self.brightness_scale.grid(row=0, column=1, padx=5, pady=5, sticky="ew")
        
        # Contrast control
        ttk.Label(filter_frame, text="Contrast:").grid(row=1, column=0, padx=5, pady=5, sticky="e")
        self.contrast_scale = ttk.Scale(
            filter_frame,
            from_=0.1,
            to=2.0,
            value=1.0,
            command=lambda v: self.adjust_contrast(float(v))
        )
        self.contrast_scale.grid(row=1, column=1, padx=5, pady=5, sticky="ew")
        
        # Saturation control
        ttk.Label(filter_frame, text="Saturation:").grid(row=2, column=0, padx=5, pady=5, sticky="e")
        self.saturation_scale = ttk.Scale(
            filter_frame,
            from_=0.0,
            to=2.0,
            value=1.0,
            command=lambda v: self.adjust_saturation(float(v))
        )
        self.saturation_scale.grid(row=2, column=1, padx=5, pady=5, sticky="ew")
        
        # Sharpen button
        sharpen_button = ttk.Button(
            filter_frame,
            text="Sharpen",
            image=self.controller.icon_images.get("sharpen_icon"),
            compound=tk.LEFT,
            command=self.sharpen_image,
            style='TButton'
        )
        sharpen_button.grid(row=3, column=0, columnspan=2, pady=5, padx=5, sticky="ew")
        sharpen_button.bind("<Enter>", lambda e: sharpen_button.config(style='Hover.TButton'))
        sharpen_button.bind("<Leave>", lambda e: sharpen_button.config(style='TButton'))
        
        # Reset button
        reset_button = ttk.Button(
            filter_frame,
            text="Reset Filters",
            command=self.reset_filters,
            style='TButton'
        )
        reset_button.grid(row=4, column=0, columnspan=2, pady=5, padx=5, sticky="ew")
        reset_button.bind("<Enter>", lambda e: reset_button.config(style='Hover.TButton'))
        reset_button.bind("<Leave>", lambda e: reset_button.config(style='TButton'))
        
        # Configure columns
        filter_frame.grid_columnconfigure(1, weight=1)

    def upload_image(self):
        """Handle image upload"""
        file_types = [
            ("Image files", "*.jpg;*.jpeg;*.png;*.bmp"),
            ("All files", "*.*")
        ]
        file_path = filedialog.askopenfilename(
            title="Select Crop Image",
            filetypes=file_types
        )
        
        if file_path:
            try:
                self.current_image_path = file_path
                self.uploaded_image = Image.open(file_path)
                self.filtered_image = self.uploaded_image.copy()
                self.display_image(self.uploaded_image)
                
                # Start processing
                self.start_processing(file_path)
                
            except Exception as e:
                self.status_label.config(text=f"Error: {str(e)}")
                logging.error(f"Image upload error: {e}")
                messagebox.showerror("Error", f"Failed to open image: {e}")

    def start_processing(self, file_path):
        """Start image processing in background thread"""
        # Reset UI
        self.result_label.config(text="Processing...")
        self.save_button.config(state=tk.DISABLED)
        self.progress.grid()
        self.progress.start()
        self.status_label.config(text="Analyzing image...")
        
        # Process in background thread
        threading.Thread(
            target=self.process_image,
            args=(file_path,),
            daemon=True
        ).start()

    def process_image(self, image_path):
        """Process image and get diagnosis"""
        try:
            # Add to recent images if not already there
            if image_path not in self.controller.recent_images:
                self.controller.recent_images.insert(0, image_path)
                if len(self.controller.recent_images) > 20:  # Limit history
                    self.controller.recent_images.pop()
            
            # Get prediction
            start_time = time.time()
            prediction = self.diagnose_disease(image_path)
            processing_time = time.time() - start_time
            
            # Update UI on main thread
            self.after(0, self.update_results, prediction, processing_time, image_path)
            
        except Exception as e:
            self.after(0, self.status_label.config, text=f"Error: {str(e)}")
            logging.error(f"Processing error: {e}")
        finally:
            self.after(0, self.progress.stop)
            self.after(0, self.progress.grid_remove)

    def update_results(self, prediction, processing_time, image_path):
        """Update UI with results"""
        self.result_label.config(text=prediction)
        self.save_button.config(state=tk.NORMAL)
        self.status_label.config(
            text=f"Analysis complete in {processing_time:.2f} seconds. Image: {os.path.basename(image_path)}"
        )
        
        # Add to diagnosis history
        self.controller.diagnosis_history.append({
            "image": image_path,
            "diagnosis": prediction,
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        })
        
        # Refresh history page
        if "HistoryPage" in self.controller.frames:
            self.controller.frames["HistoryPage"].refresh_recent_uploads()

    def display_image(self, img):
        """Display image in the preview frame"""
        try:
            # Maintain aspect ratio
            width, height = img.size
            max_size = (self.image_frame.winfo_width() - 20, 
                       self.image_frame.winfo_height() - 20)
            
            ratio = min(max_size[0]/width, max_size[1]/height)
            new_size = (int(width*ratio), int(height*ratio))
            
            img = img.resize(new_size, Image.Resampling.LANCZOS)
            img_tk = ImageTk.PhotoImage(img)
            
            self.image_label.config(image=img_tk)
            self.image_label.image = img_tk  # Keep reference
            
        except Exception as e:
            self.status_label.config(text=f"Display error: {str(e)}")
            logging.error(f"Image display error: {e}")

    def diagnose_disease(self, image_path):
        """Diagnose disease in the uploaded image"""
        if self.controller.disease_model is None:
            return "Error: Disease detection model not loaded"
        
        # First check if image is actually a crop
        if self.controller.is_non_crop_image(image_path):
            return "Error: Uploaded image doesn't appear to be a crop leaf image"
        
        try:
            # Load and preprocess the image
            img = image.load_img(image_path, target_size=(150, 150))
            img_array = image.img_to_array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)

            # Predict the disease
            predictions = self.controller.disease_model.predict(img_array)
            predicted_class = np.argmax(predictions, axis=1)[0]
            confidence = np.max(predictions) * 100
            
            # Get prediction details
            predicted_label = list(self.controller.class_labels.keys())[predicted_class]
            disease_name = " ".join(predicted_label.split("_")[1:])
            crop_type = predicted_label.split("_")[0]
            
            if confidence < 50:
                return (
                    f"Low confidence prediction ({confidence:.1f}%)\n"
                    f"Possible {crop_type} disease: {disease_name}\n"
                    "Please upload a clearer image for better results"
                )
            
            # Format results
            result = (
                f"Crop Type: {crop_type}\n"
                f"Disease: {disease_name}\n"
                f"Confidence: {confidence:.1f}%"
            )
            
            return result
            
        except Exception as e:
            logging.error(f"Diagnosis error: {e}")
            return f"Error during diagnosis: {str(e)}"

    def save_report(self):
        """Save diagnosis report as PDF"""
        if not self.current_image_path:
            messagebox.showerror("Error", "No image to generate report")
            return
            
        diagnosis_text = self.result_label.cget("text")
        default_filename = (
            f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_"
            f"{os.path.splitext(os.path.basename(self.current_image_path))[0]}_report.pdf"
        )
        
        file_path = filedialog.asksaveasfilename(
            defaultextension=".pdf",
            filetypes=[("PDF files", "*.pdf")],
            initialfile=default_filename,
            title="Save Diagnosis Report"
        )
        
        if file_path:
            try:
                # Create PDF
                pdf = FPDF()
                pdf.add_page()
                pdf.set_font("Arial", "B", 16)
                
                # Title
                pdf.cell(0, 10, "Crop Disease Diagnosis Report", ln=True, align="C")
                pdf.ln(10)
                
                # Report details
                pdf.set_font("Arial", size=12)
                pdf.cell(0, 10, f"Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", ln=True)
                pdf.cell(0, 10, f"Image: {os.path.basename(self.current_image_path)}", ln=True)
                
                # Diagnosis
                pdf.set_font("Arial", "B", 14)
                pdf.cell(0, 10, "Diagnosis Results:", ln=True)
                pdf.set_font("Arial", size=12)
                
                # Split diagnosis text into lines
                for line in diagnosis_text.split("\n"):
                    pdf.cell(0, 10, line, ln=True)
                
                # Add the image
                try:
                    # Create temporary resized image for PDF
                    img = Image.open(self.current_image_path)
                    img.thumbnail((150, 150))
                    temp_img_path = "temp_report_img.jpg"
                    img.save(temp_img_path)
                    
                    pdf.ln(10)
                    pdf.cell(0, 10, "Image Preview:", ln=True)
                    pdf.image(temp_img_path, x=50, w=100)
                    
                    # Clean up temp file
                    os.remove(temp_img_path)
                except Exception as e:
                    logging.warning(f"Couldn't embed image in PDF: {e}")
                
                # Save PDF
                pdf.output(file_path)
                messagebox.showinfo("Success", f"Report saved to:\n{file_path}")
                
            except Exception as e:
                messagebox.showerror("Error", f"Failed to save PDF: {e}")
                logging.error(f"PDF save error: {e}")

    # Image adjustment methods
    def adjust_brightness(self, factor):
        if self.uploaded_image:
            enhancer = ImageEnhance.Brightness(self.uploaded_image)
            self.filtered_image = enhancer.enhance(factor)
            self.display_image(self.filtered_image)

    def adjust_contrast(self, factor):
        if self.uploaded_image:
            enhancer = ImageEnhance.Contrast(self.uploaded_image)
            self.filtered_image = enhancer.enhance(factor)
            self.display_image(self.filtered_image)

    def adjust_saturation(self, factor):
        if self.uploaded_image:
            enhancer = ImageEnhance.Color(self.uploaded_image)
            self.filtered_image = enhancer.enhance(factor)
            self.display_image(self.filtered_image)

    def sharpen_image(self):
        if self.uploaded_image:
            self.filtered_image = self.uploaded_image.filter(ImageFilter.SHARPEN)
            self.display_image(self.filtered_image)

    def reset_filters(self):
        """Reset all image filters to original"""
        if self.uploaded_image:
            self.filtered_image = self.uploaded_image.copy()
            self.display_image(self.filtered_image)
            self.brightness_scale.set(1.0)
            self.contrast_scale.set(1.0)
            self.saturation_scale.set(1.0)


class HistoryPage(ttk.Frame):
    def __init__(self, parent, controller):
        ttk.Frame.__init__(self, parent)
        self.controller = controller

        # Configure grid
        self.grid_rowconfigure(1, weight=1)
        self.grid_columnconfigure(0, weight=1)

        # Title label
        title_label = ttk.Label(self, text="Recent Uploads", 
                             font=("Helvetica", 18, "bold"))
        title_label.grid(row=0, column=0, columnspan=4, pady=(20, 10), padx=10)

        # Listbox with scrollbar
        list_frame = ttk.Frame(self)
        list_frame.grid(row=1, column=0, columnspan=4, sticky="nsew", padx=10, pady=10)
        list_frame.grid_rowconfigure(0, weight=1)
        list_frame.grid_columnconfigure(0, weight=1)
        
        scrollbar = ttk.Scrollbar(list_frame, orient=tk.VERTICAL)
        self.recent_uploads = tk.Listbox(
            list_frame, 
            yscrollcommand=scrollbar.set,
            height=8,  
            font=("Helvetica", 12),
            selectbackground="#64B5F6"
        )
        scrollbar.config(command=self.recent_uploads.yview)
        
        self.recent_uploads.pack(side=tk.LEFT, fill=tk.BOTH, expand=True, padx=5, pady=5)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y, padx=5, pady=5)

        # Button frame
        button_frame = ttk.Frame(self)
        button_frame.grid(row=2, column=0, columnspan=4, pady=10, padx=10, sticky="ew")
        
        # Action buttons
        buttons = [
            ("View Image", "view_icon", self.view_image),
            ("Diagnose Again", "diagnose_icon", self.re_diagnose),
            ("Delete Upload", "delete_icon", self.delete_upload),
            ("Export History", "export_icon", self.export_history)
        ]
        
        for i, (text, icon_name, command) in enumerate(buttons):
            btn = ttk.Button(
                button_frame,
                text=text,
                image=controller.icon_images.get(icon_name),
                compound=tk.LEFT,
                command=command,
                style='TButton'
            )
            btn.grid(row=0, column=i, padx=5, pady=5, sticky="ew")
            btn.bind("<Enter>", lambda e, b=btn: b.config(style='Hover.TButton'))
            btn.bind("<Leave>", lambda e, b=btn: b.config(style='TButton'))
            button_frame.grid_columnconfigure(i, weight=1)

        # Back button
        back_button = ttk.Button(
            self, 
            text="Back to Home", 
            command=lambda: controller.show_frame("HomePage"), 
            style='TButton'
        )
        back_button.grid(row=3, column=0, columnspan=4, pady=10, padx=10, sticky="ew")
        back_button.bind("<Enter>", lambda e: back_button.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda e: back_button.config(style='TButton'))

    def refresh_recent_uploads(self):
        """Update the list of recent uploads"""
        self.recent_uploads.delete(0, tk.END)
        for img in self.controller.recent_images:
            self.recent_uploads.insert(tk.END, os.path.basename(img))

    def view_image(self):
        try:
            selected_index = self.recent_uploads.curselection()[0]
            image_path = self.controller.recent_images[selected_index]
            
            top = Toplevel(self)
            top.title("Image Preview")
            
            try:
                img = Image.open(image_path)
                img.thumbnail((400, 400))  # Maintain aspect ratio
                img_tk = ImageTk.PhotoImage(img)
                
                label = tk.Label(top, image=img_tk)
                label.image = img_tk  # Keep reference
                label.pack(padx=10, pady=10)
                
            except Exception as e:
                messagebox.showerror("Error", f"Failed to open image: {e}")
                top.destroy()
                
        except IndexError:
            messagebox.showerror("Error", "Please select an image first")

    def re_diagnose(self):
        try:
            selected_index = self.recent_uploads.curselection()[0]
            image_path = self.controller.recent_images[selected_index]
            
            # Switch to diagnosis page and process
            self.controller.show_frame("DiagnosisPage")
            diagnosis_page = self.controller.frames["DiagnosisPage"]
            diagnosis_page.current_image_path = image_path
            diagnosis_page.uploaded_image = Image.open(image_path)
            diagnosis_page.display_image(diagnosis_page.uploaded_image)
            
            # Start processing
            diagnosis_page.progress.grid()
            diagnosis_page.progress.start()
            threading.Thread(
                target=diagnosis_page.process_image, 
                args=(image_path,),
                daemon=True
            ).start()
            
        except IndexError:
            messagebox.showerror("Error", "Please select an image first")

    def delete_upload(self):
        try:
            selected_index = self.recent_uploads.curselection()[0]
            del self.controller.recent_images[selected_index]
            self.refresh_recent_uploads()
            
        except IndexError:
            messagebox.showerror("Error", "Please select an image to delete")

    def export_history(self):
        if not self.controller.recent_images:
            messagebox.showerror("Error", "No history to export")
            return
            
        file_path = filedialog.asksaveasfilename(
            defaultextension=".csv",
            filetypes=[("CSV files", "*.csv"), ("Text files", "*.txt")],
            title="Save Upload History"
        )
        
        if file_path:
            try:
                with open(file_path, "w") as f:
                    f.write("Filename,Date\n")
                    for img in self.controller.recent_images:
                        timestamp = time.strftime(
                            "%Y-%m-%d %H:%M", 
                            time.localtime(os.path.getmtime(img))
                        )
                        f.write(f"{os.path.basename(img)},{timestamp}\n")
                
                messagebox.showinfo("Success", f"History exported to {file_path}")
                
            except Exception as e:
                messagebox.showerror("Error", f"Failed to export: {e}")

class FeedbackPage(ttk.Frame):
    def __init__(self, parent, controller):
        ttk.Frame.__init__(self, parent)
        self.controller = controller

        # Configure grid
        self.grid_rowconfigure(1, weight=1)
        self.grid_columnconfigure(0, weight=1)

        # Title label
        title_label = ttk.Label(self, text="User Feedback", 
                             font=("Helvetica", 18, "bold")) 
        title_label.grid(row=0, column=0, columnspan=2, pady=(20, 10), padx=10)

        # Feedback text area with scrollbar
        text_frame = ttk.Frame(self)
        text_frame.grid(row=1, column=0, columnspan=2, sticky="nsew", padx=10, pady=10)
        
        scrollbar = ttk.Scrollbar(text_frame, orient=tk.VERTICAL)
        self.feedback_text = tk.Text(
            text_frame, 
            yscrollcommand=scrollbar.set,
            height=8, 
            width=50, 
            font=("Helvetica", 12), 
            padx=5, 
            pady=5,
            wrap="word"
        )
        scrollbar.config(command=self.feedback_text.yview)
        
        self.feedback_text.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

        # Submit button
        submit_button = ttk.Button(
            self,
            text="Submit Feedback",
            image=controller.icon_images.get("submit_icon"),
            compound=tk.LEFT,
            command=self.submit_feedback,
            style='TButton'
        )
        submit_button.grid(row=2, column=0, columnspan=2, pady=10, padx=10, sticky="ew")
        submit_button.bind("<Enter>", lambda e: submit_button.config(style='Hover.TButton'))
        submit_button.bind("<Leave>", lambda e: submit_button.config(style='TButton'))

        # Back button
        back_button = ttk.Button(
            self, 
            text="Back to Home", 
            command=lambda: controller.show_frame("HomePage"), 
            style='TButton'
        )
        back_button.grid(row=3, column=0, columnspan=2, pady=10, padx=10, sticky="ew")
        back_button.bind("<Enter>", lambda e: back_button.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda e: back_button.config(style='TButton'))

    def submit_feedback(self):
        feedback = self.feedback_text.get("1.0", tk.END).strip()
        if not feedback:
            messagebox.showerror("Error", "Please enter your feedback before submitting.")
            return

        # Create feedback directory if it doesn't exist
        feedback_dir = "feedback"
        os.makedirs(feedback_dir, exist_ok=True)
        
        # Save with timestamp
        timestamp = time.strftime("%Y%m%d_%H%M%S")
        file_path = os.path.join(feedback_dir, f"feedback_{timestamp}.txt")
        
        try:
            with open(file_path, "w") as f:
                f.write(feedback)
            
            messagebox.showinfo("Thank You", "Your feedback has been submitted successfully.")
            self.feedback_text.delete("1.0", tk.END)
            
        except Exception as e:
            messagebox.showerror("Error", f"Failed to save feedback: {e}")
            logging.error(f"Feedback save error: {e}")

class HelpPage(ttk.Frame):
    def __init__(self, parent, controller, icon_images=None):
        super().__init__(parent)
        self.controller = controller
        self.icon_images = {} 

        # Create canvas with scrollbar
        self.canvas = tk.Canvas(self, highlightthickness=0)
        scrollbar = ttk.Scrollbar(self, orient=tk.VERTICAL, command=self.canvas.yview)
        self.scrollable_frame = ttk.Frame(self.canvas)

        self.scrollable_frame.bind(
            "<Configure>",
            lambda e: self.canvas.configure(
                scrollregion=self.canvas.bbox("all")
            )
        )

        self.canvas.create_window((0, 0), window=self.scrollable_frame, anchor=tk.NW)
        self.canvas.configure(yscrollcommand=scrollbar.set)

        self.canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True, padx=5, pady=5)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y, padx=5, pady=5)

        # Bind mousewheel for scrolling
        self.canvas.bind_all("<MouseWheel>", self._on_mousewheel)

        # Title label
        title_label = ttk.Label(
            self.scrollable_frame, 
            text="Help & Instructions", 
            font=("Helvetica", 20, "bold"),  
        )
        title_label.pack(fill=tk.X, padx=20, pady=(20, 10))

        # Help sections
        self.help_sections = [
            {
                "title": "Getting Started",
                "icon": "help_icon",
                "content": [
                    "Welcome to the Crop Disease Diagnosis application!",
                    "This tool helps identify diseases in crop leaves using image analysis."
                ]
            },
            {
                "title": "How to Diagnose",
                "icon": "diagnose_icon",
                "content": [
                    "1. Go to the Diagnosis page from the Home screen",
                    "2. Click 'Upload Crop Image' to select an image",
                    "3. The system will analyze the image and show results",
                    "4. Use filters if needed to improve image quality",
                    "5. Save the diagnosis report when complete"
                ]
            },
            {
                "title": "Viewing History",
                "icon": "history_icon",
                "content": [
                    "• View your previously uploaded images",
                    "• Re-run diagnosis on old images",
                    "• Delete images from your history",
                    "• Export your history to a CSV file"
                ]
            },
            {
                "title": "Providing Feedback",
                "icon": "feedback_icon",
                "content": [
                    "We welcome your feedback to improve the application!",
                    "Use the Feedback page to share your thoughts."
                ]
            },
            {
                "title": "Dark Mode",
                "icon": "dark_mode_icon",
                "content": [
                    "Toggle between light and dark color schemes",
                    "Easier on the eyes in low-light conditions"
                ]
            },
            {
                "title": "Need More Help?",
                "icon": "help_icon",
                "content": [
                    "Contact support at: ayushchoubey800@gmail.com",
                    "We're happy to assist with any questions!"
                ]
            }
        ]

        # Create help sections
        for section in self.help_sections:
            self.create_section(section)

        # Back button
        back_button = ttk.Button(
            self.scrollable_frame,
            text="Back to Home",
            command=lambda: controller.show_frame("HomePage"),
            style='TButton'
        )
        back_button.pack(pady=20, ipady=5, ipadx=20)
        back_button.bind("<Enter>", lambda e: back_button.config(style='Hover.TButton'))
        back_button.bind("<Leave>", lambda e: back_button.config(style='TButton'))

    def create_section(self, section):
        """Creates a help section with icon, title and content"""
        section_frame = ttk.Frame(self.scrollable_frame)
        section_frame.pack(fill=tk.X, padx=20, pady=10)

        # Icon
        icon = self.icon_images.get(section["icon"])
        if icon:
            icon_label = ttk.Label(section_frame, image=icon)
            icon_label.image = icon  # Keep reference
            icon_label.pack(side=tk.LEFT, padx=(0, 10))

        # Content frame
        content_frame = ttk.Frame(section_frame)
        content_frame.pack(side=tk.LEFT, fill=tk.X, expand=True)

        # Title
        title_label = ttk.Label(
            content_frame, 
            text=section["title"], 
            font=("Helvetica", 14, "bold"), 
            anchor=tk.W
        )
        title_label.pack(fill=tk.X)

        # Content items
        for item in section["content"]:
            content_label = ttk.Label(
                content_frame,
                text=item,
                font=("Helvetica", 12),
                justify="left",
                anchor=tk.W,
                wraplength=600
            )
            content_label.pack(fill=tk.X, pady=2)

        # Separator
        ttk.Separator(self.scrollable_frame, orient=tk.HORIZONTAL).pack(fill=tk.X, padx=20, pady=10)

    def _on_mousewheel(self, event):
        """Handle mousewheel scrolling"""
        self.canvas.yview_scroll(int(-1*(event.delta/120)), "units")

if __name__ == "__main__":
    try:
        app = App()
        app.mainloop()
    except Exception as e:
        logging.error(f"Application error: {e}")
        messagebox.showerror("Fatal Error", f"The application encountered an error and will close:\n{str(e)}")
```
